Linear Function- plotting using **plot()** function of **pyplot** module of the **matplotlib** library                                                       
plot(x=label,y=label,kind='type')                                                         
kind- to set the type of plot we want                                                
-line: This is the default type of plot and it creates a line
plot.                             
-bar: This creates a bar chart.                                                                
-scatter: This creates a scatter plot.                                                        
-hist: This creates a histogram.                                                          
-box: This creates a box plot.                                                              
-kde: This creates a kernel density estimation plot.                                            
-pie: This creates a pie chart.                                                           
-polar: This creates a polar plot.                                                            
-3d: This creates a 3D plot        

```python
import pandas as pd
hd=pd.read_csv("SportsWatch_St.csv",header=0,sep=",")
```

```python
import matplotlib.pyplot as plt
hd.plot(x="Average_Pulse",y="Calorie_Burnage",kind='scatter')
plt.xlim(xmin=0)
plt.ylim(ymin=0)
plt.show()
```

⏩ **Finding slope**                                                                                                                                          
 using the user-defined function

```python
def slope(x1,y1,x2,y2):
  s=(y2-y1)/(x2-x1)
  return s
print(slope(80,240,85,250))
```

⏩**Finding both slope and intercept**- using the **polyfit()** function of **numpy** library

```python
import numpy as np
x=hd["Average_Pulse"]
y=hd["Calorie_Burnage"]
si=np.polyfit(x,y,1)
print(si)
```
